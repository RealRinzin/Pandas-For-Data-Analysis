#### Read CSV File from Internet

```python
import pandas as pd
# read_csv_file = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',verify_ssl=False)
read_csv_file = pd.read_csv('./crime_data.csv')
read_csv_file.head(6)
```

#### Exporting the data to CSV Format

```python
# List of list
list = [[1,"Rahul","M","A+"],[2,"Tenzin","M","A"]]

# Converting it to DataFrame
df = pd.DataFrame(list,columns=['ID','name','gender','grade'])

# Exporting it to the CSV format
df.to_csv('exported_to_format.csv')
```

#### Describe method to overview the statistical values of the table 

```python
# Describe usually return the min,max,count,avg of all the numerical columns
# but if you want it for all the columns then you have to 
# pass include='all' parameter in describe function
read_csv_file.describe()
```

```python
# with include='all' parameter | usually it does not make lots of sense to call non-numerical columns for statistical overview
read_csv_file.describe(include='all')
```
